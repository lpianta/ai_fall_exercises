```
# loading libraries

import requests
from bs4 import BeautifulSoup
import re
```

```
# getting page

page = requests.get('https://www.goodreads.com/list/show/6.Best_Books_of_the_20th_Century')
soup = BeautifulSoup(page.content, 'html.parser')
```

```
# page turner


link_base = 'https://www.goodreads.com/list/show/6.Best_Books_of_the_20th_Century?page={}'

page_list = []
for index in range(1, 3):
    link_string = link_base.format(index)
    page_list.append(link_string)
print(page_list)

# getting individual book link

link_list = []
for i in page_list:
    new_page = requests.get(i)
    new_soup = BeautifulSoup(new_page.content, 'html.parser')
    books_links = new_soup.find_all('a', class_='bookTitle', href=True)
    for i in books_links:
        href = i['href']
        link = 'https://www.goodreads.com' + href
        link_list.append(link)
print(len(link_list))
```

```
# initializing empty lists
num_rating_list = []
avg_rating_list = []
num_reviews_list = []

for i in link_list:
    another_page = requests.get(i)
    another_soup = BeautifulSoup(another_page.content, 'html.parser')
    # finding avg rating and updating list
    rating = another_soup.find('span', itemprop="ratingValue").get_text()
    avg_rating_list.append(rating)
    # finding ratings count, cleaning data and updating list
    rating_count = re.sub('[\D+]', '', another_soup.find('meta', itemprop = "ratingCount").text)
    num_rating_list.append(rating_count)
    # finding review count, cleaning data and updating list
    review_count = re.sub('[\D+]', '', another_soup.find('meta', itemprop = "reviewCount").text)
    num_reviews_list.append(review_count)
    
```
