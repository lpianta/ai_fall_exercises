# Exercise 3: Get the city

```python
def get_city(name):
    if name.split()[0].strip() != '!':
        return name.split()[0].strip()
    else:
        return name.split()[1].strip()

shop['city'] = shop.shop_name.map(get_city)

shop['city'].loc[shop['city'] == 'St.'] = 'St. Petersburg'
shop['city'].loc[shop['city'] == 'RostovNaDon'] = 'Rostov Na Don'
shop['city'].loc[shop['city'] == 'RostovNaDonu'] = 'Rostov Na Don'

shop
```

# Exercise 2: Pysical vs Online shop

```python
def online(shop):
    if shop == 'Online':
        return 1
    elif shop == 'Digital':
        return 1
    elif shop == 'Internet-shop':
        return 1
    else:
        return 0
    
shop['online'] = shop['city'].map(online)

shop.head()
```

# Exercise 4: Latitude & Longitude

```python
shop['city'] = shop['city'].astype(str)
```

```python
geolocator = Nominatim(user_agent='app')
```

```python
def get_lat(city):
    if geolocator.geocode(city) is not 'None':
        return geolocator.geocode(city).latitude
    else:
        return 'None'
shop['latitude'] = shop['city'].map(get_lat)
```

```python
def get_long(city):
    if geolocator.geocode(city) is not 'None':
        return geolocator.geocode(city).longitude
    else:
        return 'None'
shop['longitude'] = shop['city'].map(get_long)
```
