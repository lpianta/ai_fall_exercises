# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Have a look at the 2 images below
text = cv2.imread('img/t1.png')
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray')
```

```python
text_2 = cv2.imread('img/t2.png')
plt.figure(figsize = (12,10))
plt.imshow(text_2, cmap='gray')
```

```python
# Apply threshold to the images above to make them look better, try to find the best one

text_copy = text.copy()

fig, axes = plt.subplots(2, 3, figsize=(20, 10))

text_bw = cv2.cvtColor(text_copy, cv2.COLOR_BGR2GRAY)
ret, text_binary = cv2.threshold(text_bw, 127, 255, cv2.THRESH_BINARY)
ret, text_binary_inv = cv2.threshold(text_bw, 127, 255, cv2.THRESH_BINARY_INV)
ret, text_trunc = cv2.threshold(text_bw, 127, 255, cv2.THRESH_TRUNC)
ret, text_tozero = cv2.threshold(text_bw, 127, 255, cv2.THRESH_TOZERO)
ret, text_tozero_inv = cv2.threshold(text_bw, 127, 255, cv2.THRESH_TOZERO_INV)


axes[0,0].imshow(text_bw, cmap='gray')
axes[0,1].imshow(text_binary, cmap='gray')
axes[0,2].imshow(text_binary_inv, cmap='gray')
axes[1,0].imshow(text_trunc, cmap='gray')
axes[1,1].imshow(text_tozero, cmap='gray')
axes[1,2].imshow(text_tozero_inv, cmap='gray')

axes[0,0].title.set_text('bw')
axes[0,1].title.set_text('binary')
axes[0,2].title.set_text('binary_inv')
axes[1,0].title.set_text('trunc')
axes[1,1].title.set_text('tozero')
axes[1,2].title.set_text('tozero_inv')
```

```python
# Apply threshold to the images above to make them look better, try to find the best one

text2_copy = text_2.copy()

fig, axes = plt.subplots(2, 3, figsize=(20, 10))

text2_bw = cv2.cvtColor(text2_copy, cv2.COLOR_BGR2GRAY)
ret, text2_binary = cv2.threshold(text2_bw, 127, 255, cv2.THRESH_BINARY)
ret, text2_binary_inv = cv2.threshold(text2_bw, 127, 255, cv2.THRESH_BINARY_INV)
ret, text2_trunc = cv2.threshold(text2_bw, 127, 255, cv2.THRESH_TRUNC)
ret, text2_tozero = cv2.threshold(text2_bw, 127, 255, cv2.THRESH_TOZERO)
ret, text2_tozero_inv = cv2.threshold(text2_bw, 127, 255, cv2.THRESH_TOZERO_INV)


axes[0,0].imshow(text2_bw, cmap='gray')
axes[0,1].imshow(text2_binary, cmap='gray')
axes[0,2].imshow(text2_binary_inv, cmap='gray')
axes[1,0].imshow(text2_trunc, cmap='gray')
axes[1,1].imshow(text2_tozero, cmap='gray')
axes[1,2].imshow(text2_tozero_inv, cmap='gray')

axes[0,0].title.set_text('bw')
axes[0,1].title.set_text('binary')
axes[0,2].title.set_text('binary_inv')
axes[1,0].title.set_text('trunc')
axes[1,1].title.set_text('tozero')
axes[1,2].title.set_text('tozero_inv')
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C

mean_thresh = cv2.adaptiveThreshold(text_bw, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 9)

plt.figure(figsize=(10,10))
plt.imshow(mean_thresh, cmap='gray')
```

```python
mean_thresh_2 = cv2.adaptiveThreshold(text2_bw, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 29, 4)

plt.figure(figsize=(10,10))
plt.imshow(mean_thresh_2, cmap='gray')
```

```python
gauss_thresh = cv2.adaptiveThreshold(text_bw, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 4)

plt.figure(figsize=(10,10))
plt.imshow(gauss_thresh, cmap='gray')
```

```python
gauss_thresh = cv2.adaptiveThreshold(text2_bw, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 23, 3)

plt.figure(figsize=(10,10))
plt.imshow(gauss_thresh, cmap='gray')
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image

bw_copy = text_bw.copy()

ret, otsu = cv2.threshold(bw_copy, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(10,10))
plt.imshow(otsu, cmap='gray')
```

```python
bw_2_copy = text2_bw.copy()

ret, otsu_2 = cv2.threshold(bw_2_copy, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(10,10))
plt.imshow(otsu_2, cmap='gray')
```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding

bw_blur = cv2.GaussianBlur(bw_copy, (5,5), 0)
ret, blur_otsu = cv2.threshold(bw_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(10,10))
plt.imshow(blur_otsu, cmap='gray')
```

```python
bw_blur_2 = cv2.GaussianBlur(bw_2_copy, (5,5), 0)
ret, blur_otsu_2 = cv2.threshold(bw_blur_2, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(10,10))
plt.imshow(blur_otsu_2, cmap='gray')
```
