# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Histograms

```python
def imgshow(img):
    plt.figure(figsize=(20,20))
    plt.imshow((img[:,:,::-1]))
```

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc

path = './img/'

dog = cv2.imread(path + 'dog.jpg')
cat = cv2.imread(path + 'cat.jpg')
bottle = cv2.imread(path + 'bottiglia.jpg')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 15))



dog_b_h = cv2.calcHist([dog], [0], None, [255], [0,255])
axes[0].plot(dog_b_h, color='blue')


dog_g_h = cv2.calcHist([dog], [1], None, [255], [0,255])
axes[1].plot(dog_g_h, color='green')

dog_r_h = cv2.calcHist([dog], [2], None, [255], [0,255])
axes[2].plot(dog_r_h, color='red')

```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?


plt.figure(figsize=(20,8))

colors = ['b', 'g', 'r']

for i, colors in enumerate(colors):
    histogram = cv2.calcHist([dog], [i], None, [255], [0,255])
    plt.plot(histogram, color=colors)
    plt.xlim(0,255)

plt.figure(figsize=(20,10))
plt.show()
```

```python
def day_night(imgpath):
    img = cv2.imread(imgpath)
    img_copy = img.copy()
    img_hsv = cv2.cvtColor(img_copy, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(img_hsv)
    if v.mean() > 100:
        print('Daytime')
    else:
        print('Nighttime')
        
day_night(path + 'night4.jpg')
```
