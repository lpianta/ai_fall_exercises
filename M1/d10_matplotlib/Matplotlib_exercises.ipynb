### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
print(data)
fig = plt.figure()
plt.plot(data['month_number'], data['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.show()

```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here
fig = plt.figure(figsize=(10,8))
plt.plot(data['month_number'], data['total_profit'], color='red', linestyle=':', linewidth='3', marker='o', markerfacecolor='black')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.legend('profit', loc='lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

```python
data.head()
```

```python
# Write your code here
fig, ax = plt.subplots(figsize=(10,8))
facecream = plt.plot(data['month_number'], data['facecream'], marker='o')
facewash = plt.plot(data['month_number'], data['facewash'], marker='o')
toothpaste = plt.plot(data['month_number'], data['toothpaste'], marker='o')
bathingsoap = plt.plot(data['month_number'], data['bathingsoap'], marker='o')
shampoo = plt.plot(data['month_number'], data['shampoo'], marker='o')
moisturizer = plt.plot(data['month_number'], data['moisturizer'], marker='o')
plt.legend(['facecream sales data', 'facewash sales data', 'toothpaste sales data', 'bathingsoap sales data', 'shampoo sales data', 'moisturizer sales data'])

```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.
The histogram should look like this:


```python
# Write your code here
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(data['total_profit'], bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000])
ax.set_title('Profit Data')
ax.set_ylabel('Actual profit in dollars')
ax.set_xlabel('Profit range in dollars')
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here

sums = data.sum(axis='index')
facecream_sum = sums[1]
facewash_sum = sums[2]
toothpaste_sum = sums[3]
bathingsoap_sum = sums[4]
shampoo_sum = sums[5]
moisturizer_sum = sums[6]

pie_data = [facecream_sum, facewash_sum, toothpaste_sum, bathingsoap_sum, shampoo_sum, moisturizer_sum]

pie_label = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

fig, ax = plt.subplots(figsize=(10, 10))

ax.pie(pie_data, labels=pie_label, autopct='%1.1f%%')

plt.legend(loc='lower right')
plt.show()
```
